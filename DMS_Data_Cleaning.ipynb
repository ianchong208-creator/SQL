```python
import pandas as pd

# 1. Define columns
columns = [
    'InvoiceNo', 'StockCode', 'Description', 'Quantity',
    'InvoiceDate', 'UnitPrice', 'CustomerID', 'Country'
]

# 2. Load the CSV with the correct encoding
# 'ISO-8859-1' handles the special characters (like Â£) causing the crash
df = pd.read_csv('data.csv', header=None, names=columns, encoding='ISO-8859-1')

# 3. Fix Date Format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%Y %H:%M')
df['InvoiceDate'] = df['InvoiceDate'].dt.strftime('%Y-%m-%d %H:%M:%S')

# 4. Handle Missing Customer IDs
df['CustomerID'] = df['CustomerID'].fillna('')

# 5. Save as UTF-8 (Standard for Cassandra)
df.to_csv('cleaned_online_retail.csv', index=False, encoding='utf-8')


```
